```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data
df = pd.read_csv('../responses.csv')

# Filter out non-engineering teams
teams_to_exclude = ['Customer Solutions', 'Customer Support', 'Pre-Sales']
df_filtered = df[~df['Which team are you a part of?'].isin(teams_to_exclude)].copy()

# Get column names
team_col = 'Which team are you a part of?'
helpfulness_col = 'On a scale of 1 to 5, how would you rate the overall helpfulness of AI in your daily development tasks?'
agents_col = df_filtered.columns[16]  # Column Q

# Create numeric mapping for helpfulness
rating_values = {
    'Extremely Helpful': 5,
    'Very Helpful': 4,
    'Moderately Helpful': 3,
    'Slightly Helpful': 2,
    'Not Helpful': 1
}

df_filtered['rating_numeric'] = df_filtered[helpfulness_col].map(rating_values)

# Define agents.md order
agents_order = [
    'Yes -- and it made a meaningful difference',
    'Yes -- but I didn\'t notice much impact',
    'No -- I know it\'s possible but haven\'t taken the time',
    'No -- I wasn\'t aware that was possible'
]

# Shortened labels for display
agents_labels = [
    'Made meaningful\ndifference',
    'No noticeable\nimpact',
    'Haven\'t tried\nyet',
    'Not aware'
]

def create_team_correlation_chart(team_name, df_filtered):
    """Create a chart showing correlation between agents.md usage and helpfulness for a specific team"""
    
    # Filter data for this team
    team_df = df_filtered[df_filtered[team_col] == team_name].copy()
    
    if len(team_df) == 0:
        print(f"No data found for team: {team_name}")
        return
    
    # Calculate average helpfulness by agents.md response
    avg_by_agents = team_df.groupby(agents_col)['rating_numeric'].agg(['mean', 'count'])
    avg_by_agents = avg_by_agents.reindex(agents_order)
    
    # Calculate overall team average
    team_avg = team_df['rating_numeric'].mean()
    
    # Create figure
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Define colors - modern teal/blue gradient for positive, orange/red for negative
    colors = ['#0D9488', '#14B8A6', '#FBBF24', '#EF4444']
    
    # Create bar chart
    x_pos = np.arange(len(agents_order))
    bars = []
    for i, (response, color, label) in enumerate(zip(agents_order, colors, agents_labels)):
        if response in avg_by_agents.index and not pd.isna(avg_by_agents.loc[response, 'mean']):
            avg = avg_by_agents.loc[response, 'mean']
            count = int(avg_by_agents.loc[response, 'count'])
            bar = ax.bar(i, avg, color=color, edgecolor='white', linewidth=2, width=0.6)
            bars.append(bar)
            
            # Add value label
            ax.text(i, avg + 0.1, f'{avg:.2f}\n(n={count})', 
                   ha='center', va='bottom', fontsize=11, weight='bold', color='#1F2937')
        else:
            # No data for this category
            bar = ax.bar(i, 0, color='#CCCCCC', edgecolor='white', linewidth=2, width=0.6, alpha=0.3)
            bars.append(bar)
            ax.text(i, 0.1, 'No data', ha='center', va='bottom', fontsize=9, style='italic', color='#666666')
    
    # Add team average line - use teal for consistency
    ax.axhline(team_avg, color='#0D9488', linestyle='--', linewidth=2.5, 
              alpha=0.8, label=f'Team Avg: {team_avg:.2f}')
    
    # Customize plot
    ax.set_xticks(x_pos)
    ax.set_xticklabels(agents_labels, fontsize=10, weight='bold')
    ax.set_ylabel('Average Helpfulness Rating', fontsize=12, weight='bold')
    ax.set_title(f'{team_name} Team: AI Helpfulness vs Advanced Configuration Usage', 
                fontsize=16, weight='bold', pad=20)
    ax.set_ylim(0, 5.5)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.legend(fontsize=11, loc='upper right')
    ax.grid(axis='y', alpha=0.3, linestyle='--')
    
    # Add rating scale reference
    ax.text(0.02, 0.98, '5 = Extremely Helpful\n4 = Very Helpful\n3 = Moderately Helpful', 
            transform=ax.transAxes, ha='left', va='top',
            fontsize=9, style='italic', color='#666666',
            bbox=dict(boxstyle='round,pad=0.5', facecolor='#F5F5F5', alpha=0.8))
    
    # Add subtitle with team size
    plt.text(0.5, -0.12, f'n = {len(team_df)} team members', 
            ha='center', transform=ax.transAxes,
            fontsize=10, style='italic', color='#666666')
    
    plt.tight_layout()
    plt.show()
    
    # Print summary for this team
    print(f"\n{'='*80}")
    print(f"{team_name.upper()} TEAM - ADVANCED CONFIG vs HELPFULNESS CORRELATION")
    print(f"{'='*80}")
    
    print(f"\n Team Size: {len(team_df)} members")
    print(f"   Team Average Helpfulness: {team_avg:.2f}/5.0")
    
    print(f"\n Helpfulness by Advanced Configuration Usage:")
    for response, label in zip(agents_order, agents_labels):
        if response in avg_by_agents.index and not pd.isna(avg_by_agents.loc[response, 'mean']):
            avg = avg_by_agents.loc[response, 'mean']
            count = int(avg_by_agents.loc[response, 'count'])
            diff = avg - team_avg
            comparison = f"(+{diff:.2f})" if diff > 0 else f"({diff:.2f})"
            print(f"    {response}")
            print(f"     Average: {avg:.2f}/5.0 (n={count}) {comparison} vs team avg")
        else:
            print(f"    {response}")
            print(f"     No data")
    
    # Calculate insights
    tried_configs = team_df[team_df[agents_col].isin(agents_order[:2])]
    if len(tried_configs) > 0:
        tried_avg = tried_configs['rating_numeric'].mean()
        not_tried = team_df[team_df[agents_col].isin(agents_order[2:])]
        not_tried_avg = not_tried['rating_numeric'].mean() if len(not_tried) > 0 else 0
        
        print(f"\n Key Insights:")
        print(f"    Engineers who tried advanced config rate AI at {tried_avg:.2f}/5.0")
        if not_tried_avg > 0:
            print(f"    Engineers who haven't tried rate AI at {not_tried_avg:.2f}/5.0")
            print(f"    Difference: {tried_avg - not_tried_avg:.2f} points")
        
        helpful_count = len(team_df[team_df[agents_col] == agents_order[0]])
        print(f"    {helpful_count} team members found advanced config made a meaningful difference")
    
    print(f"\n{'='*80}\n")


def create_all_teams_correlation_chart(df_filtered):
    """Create a chart showing correlation between agents.md usage and helpfulness across ALL engineering teams"""
    
    # Calculate average helpfulness by agents.md response for all teams
    avg_by_agents = df_filtered.groupby(agents_col)['rating_numeric'].agg(['mean', 'count'])
    avg_by_agents = avg_by_agents.reindex(agents_order)
    
    # Calculate overall average
    overall_avg = df_filtered['rating_numeric'].mean()
    
    # Create figure
    fig, ax = plt.subplots(figsize=(10, 6.4))
    
    # Define colors - modern teal/blue gradient for positive, orange/red for negative
    colors = ['#0D9488', '#14B8A6', '#FBBF24', '#EF4444']
    
    # Create bar chart
    x_pos = np.arange(len(agents_order))
    bars = []
    for i, (response, color, label) in enumerate(zip(agents_order, colors, agents_labels)):
        if response in avg_by_agents.index and not pd.isna(avg_by_agents.loc[response, 'mean']):
            avg = avg_by_agents.loc[response, 'mean']
            count = int(avg_by_agents.loc[response, 'count'])
            bar = ax.bar(i, avg, color=color, edgecolor='white', linewidth=2, width=0.6)
            bars.append(bar)
            
            # Add value label
            ax.text(i, avg + 0.1, f'{avg:.2f}\n(n={count})', 
                   ha='center', va='bottom', fontsize=11, weight='bold', color='#1F2937')
        else:
            # No data for this category
            bar = ax.bar(i, 0, color='#CCCCCC', edgecolor='white', linewidth=2, width=0.6, alpha=0.3)
            bars.append(bar)
            ax.text(i, 0.1, 'No data', ha='center', va='bottom', fontsize=9, style='italic', color='#666666')
    
    # Add overall average line - use teal for consistency
    ax.axhline(overall_avg, color='#0D9488', linestyle='--', linewidth=2.5, 
              alpha=0.8, label=f'Overall Avg: {overall_avg:.2f}')
    
    # Customize plot
    ax.set_xticks(x_pos)
    ax.set_xticklabels(agents_labels, fontsize=10, weight='bold')
    ax.set_ylabel('Average Helpfulness Rating', fontsize=12, weight='bold')
    ax.set_title('All Engineering Teams: AI Helpfulness vs Advanced Configuration Usage', 
                fontsize=16, weight='bold', pad=20)
    ax.set_ylim(0, 5.5)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.legend(fontsize=11, loc='upper right')
    ax.grid(axis='y', alpha=0.3, linestyle='--')
    
    # Add rating scale reference
    ax.text(0.02, 0.98, '5 = Extremely Helpful\n4 = Very Helpful\n3 = Moderately Helpful', 
            transform=ax.transAxes, ha='left', va='top',
            fontsize=9, style='italic', color='#666666',
            bbox=dict(boxstyle='round,pad=0.5', facecolor='#F5F5F5', alpha=0.8))
    
    # Add subtitle with total engineers
    plt.text(0.5, -0.12, f'n = {len(df_filtered)} engineers across all teams', 
            ha='center', transform=ax.transAxes,
            fontsize=10, style='italic', color='#666666')
    
    plt.tight_layout()
    plt.show()
    
    # Print summary for all teams
    print(f"\n{'='*80}")
    print(f"ALL ENGINEERING TEAMS - ADVANCED CONFIG vs HELPFULNESS CORRELATION")
    print(f"{'='*80}")
    
    print(f"\n Total Engineers: {len(df_filtered)}")
    print(f"   Overall Average Helpfulness: {overall_avg:.2f}/5.0")
    
    print(f"\n Helpfulness by Advanced Configuration Usage (All Teams):")
    for response, label in zip(agents_order, agents_labels):
        if response in avg_by_agents.index and not pd.isna(avg_by_agents.loc[response, 'mean']):
            avg = avg_by_agents.loc[response, 'mean']
            count = int(avg_by_agents.loc[response, 'count'])
            diff = avg - overall_avg
            comparison = f"(+{diff:.2f})" if diff > 0 else f"({diff:.2f})"
            print(f"    {response}")
            print(f"     Average: {avg:.2f}/5.0 (n={count}) {comparison} vs overall avg")
        else:
            print(f"    {response}")
            print(f"     No data")
    
    # Calculate insights
    tried_configs = df_filtered[df_filtered[agents_col].isin(agents_order[:2])]
    if len(tried_configs) > 0:
        tried_avg = tried_configs['rating_numeric'].mean()
        not_tried = df_filtered[df_filtered[agents_col].isin(agents_order[2:])]
        not_tried_avg = not_tried['rating_numeric'].mean() if len(not_tried) > 0 else 0
        
        print(f"\n Key Insights:")
        print(f"    Engineers who tried advanced config rate AI at {tried_avg:.2f}/5.0")
        if not_tried_avg > 0:
            print(f"    Engineers who haven't tried rate AI at {not_tried_avg:.2f}/5.0")
            print(f"    Difference: {tried_avg - not_tried_avg:.2f} points")
        
        helpful_count = len(df_filtered[df_filtered[agents_col] == agents_order[0]])
        tried_count = len(tried_configs)
        helpful_pct = (helpful_count / tried_count * 100) if tried_count > 0 else 0
        
        print(f"    {helpful_count} engineers found advanced config made a meaningful difference")
        print(f"    That's {helpful_pct:.0f}% of those who tried advanced configuration")
        
        # Calculate potential impact
        not_tried_count = len(not_tried)
        print(f"    {not_tried_count} engineers haven't tried advanced config yet")
        print(f"    If they achieved similar results, average helpfulness could increase")
    
    print(f"\n{'='*80}\n")


# Get list of engineering teams
teams = df_filtered[team_col].unique()
teams = sorted([t for t in teams if pd.notna(t)])

print(f"\nGenerating correlation charts for {len(teams)} engineering teams...")
print(f"Teams: {', '.join(teams)}\n")

# Generate chart for each team
for team in teams:
    create_team_correlation_chart(team, df_filtered)

# Generate chart for all teams combined
print("\n" + "="*80)
print("GENERATING OVERALL SUMMARY CHART FOR ALL TEAMS")
print("="*80 + "\n")

create_all_teams_correlation_chart(df_filtered)
```
