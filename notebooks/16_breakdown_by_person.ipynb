```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data
df = pd.read_csv('../responses.csv')

# Filter out non-engineering teams
teams_to_exclude = ['Customer Solutions', 'Customer Support', 'Pre-Sales']
df_filtered = df[~df['Which team are you a part of?'].isin(teams_to_exclude)].copy()

# Get column names
name_col = 'Name'
team_col = 'Which team are you a part of?'
tech_stack_col = 'Where do you primarily work in the tech stack'
workflow_col = df_filtered.columns[14]  # Column O - Workflows
helpfulness_col = df_filtered.columns[15]  # Column P - Helpfulness

print(f"Creating individual charts for {len(df_filtered)} engineers...")

# Create numeric mapping for helpfulness
rating_values = {
    'Extremely Helpful': 5,
    'Very Helpful': 4,
    'Moderately Helpful': 3,
    'Slightly Helpful': 2,
    'Not Helpful': 1
}

# Define all possible workflows (we'll get this from the data)
all_workflows_set = set()
for response in df_filtered[workflow_col].dropna():
    workflows = [wf.strip() for wf in response.split(',')]
    all_workflows_set.update(workflows)

all_workflows = sorted(list(all_workflows_set))

def create_individual_profile(row, all_workflows):
    """Create an individual profile chart for one engineer"""
    
    name = row[name_col] if pd.notna(row[name_col]) else "Anonymous"
    team = row[team_col] if pd.notna(row[team_col]) else "Unknown"
    tech_stack = row[tech_stack_col] if pd.notna(row[tech_stack_col]) else "Unknown"
    helpfulness_rating = row[helpfulness_col] if pd.notna(row[helpfulness_col]) else "Not provided"
    workflow_response = row[workflow_col]
    
    # Parse workflows
    person_workflows = []
    if pd.notna(workflow_response):
        person_workflows = [wf.strip() for wf in workflow_response.split(',')]
    
    # Create binary array for workflows (1 if they use it, 0 if not)
    workflow_usage = [1 if wf in person_workflows else 0 for wf in all_workflows]
    
    # Get numeric rating
    rating_numeric = rating_values.get(helpfulness_rating, 0)
    
    # Create figure with custom layout
    fig = plt.figure(figsize=(12, 8))
    gs = fig.add_gridspec(3, 2, hspace=0.4, wspace=0.3, height_ratios=[1, 3, 1])
    
    # ===== TOP: Header with person info (spans both columns) =====
    ax_header = fig.add_subplot(gs[0, :])
    ax_header.axis('off')
    
    # Determine rating color
    if rating_numeric >= 4.5:
        rating_color = '#1B5E20'
        rating_bg = '#E8F5E9'
    elif rating_numeric >= 3.5:
        rating_color = '#43A047'
        rating_bg = '#F1F8E9'
    elif rating_numeric >= 2.5:
        rating_color = '#FFA726'
        rating_bg = '#FFF8E1'
    else:
        rating_color = '#E74C3C'
        rating_bg = '#FFEBEE'
    
    header_text = f"""
    ENGINEER: {name}
    Team: {team} | Tech Stack: {tech_stack}
    """
    
    ax_header.text(0.5, 0.5, header_text, 
                   transform=ax_header.transAxes,
                   ha='center', va='center',
                   fontsize=14, weight='bold',
                   bbox=dict(boxstyle='round,pad=1', facecolor='#F5F5F5', 
                            edgecolor='#2C3E50', linewidth=2))
    
    # ===== MIDDLE LEFT: Workflow Usage =====
    ax_workflows = fig.add_subplot(gs[1, 0])
    
    # Create horizontal bar chart
    y_pos = np.arange(len(all_workflows))
    colors = ['#2E7D32' if used else '#E0E0E0' for used in workflow_usage]
    
    bars = ax_workflows.barh(y_pos, workflow_usage, color=colors, 
                             height=0.7, edgecolor='#424242', linewidth=1)
    
    ax_workflows.set_yticks(y_pos)
    ax_workflows.set_yticklabels(all_workflows, fontsize=9)
    ax_workflows.set_xlim(-0.2, 1.5)
    ax_workflows.set_xticks([0, 1])
    ax_workflows.set_xticklabels(['Not Used', 'Used'], fontsize=10, weight='bold')
    ax_workflows.set_title('AI Usage Across Workflow Stages', fontsize=13, weight='bold', pad=10)
    ax_workflows.spines['top'].set_visible(False)
    ax_workflows.spines['right'].set_visible(False)
    ax_workflows.spines['bottom'].set_visible(False)
    
    # Add checkmarks and X's
    for i, used in enumerate(workflow_usage):
        if used:
            ax_workflows.text(1.1, i, 'âœ“', fontsize=16, weight='bold', 
                            color='#1B5E20', va='center')
        else:
            ax_workflows.text(0.5, i, 'âœ—', fontsize=12, 
                            color='#BDBDBD', va='center', ha='center')
    
    # ===== MIDDLE RIGHT: Helpfulness Rating =====
    ax_rating = fig.add_subplot(gs[1, 1])
    ax_rating.axis('off')
    
    # Create a large display of the rating
    rating_display = f"""
    
    
    AI HELPFULNESS RATING
    
    
    {helpfulness_rating}
    
    
    {rating_numeric}/5.0
    
    """
    
    ax_rating.text(0.5, 0.5, rating_display,
                  transform=ax_rating.transAxes,
                  ha='center', va='center',
                  fontsize=18, weight='bold',
                  bbox=dict(boxstyle='round,pad=1.5', facecolor=rating_bg,
                           edgecolor=rating_color, linewidth=4),
                  color=rating_color)
    
    # Add rating scale reference
    scale_text = "5 = Extremely Helpful\n4 = Very Helpful\n3 = Moderately Helpful\n2 = Slightly Helpful\n1 = Not Helpful"
    ax_rating.text(0.5, 0.05, scale_text,
                  transform=ax_rating.transAxes,
                  ha='center', va='bottom',
                  fontsize=8, style='italic', color='#666666')
    
    # ===== BOTTOM: Summary Stats (spans both columns) =====
    ax_summary = fig.add_subplot(gs[2, :])
    ax_summary.axis('off')
    
    workflows_used_count = sum(workflow_usage)
    total_workflows = len(all_workflows)
    coverage_pct = (workflows_used_count / total_workflows * 100) if total_workflows > 0 else 0
    
    summary_text = f"Workflow Coverage: {workflows_used_count}/{total_workflows} stages ({coverage_pct:.0f}%)  |  "
    
    if workflows_used_count >= 6:
        summary_text += "âœ“ Comprehensive AI integration across development lifecycle"
    elif workflows_used_count >= 4:
        summary_text += "Good workflow integration - opportunity to expand to more stages"
    elif workflows_used_count >= 2:
        summary_text += "Moderate usage - significant opportunity for broader integration"
    else:
        summary_text += "Limited usage - focus on identifying high-value workflows to start"
    
    ax_summary.text(0.5, 0.5, summary_text,
                   transform=ax_summary.transAxes,
                   ha='center', va='center',
                   fontsize=11, weight='bold',
                   bbox=dict(boxstyle='round,pad=0.8', facecolor='#E3F2FD',
                            edgecolor='#1976D2', linewidth=2))
    
    plt.tight_layout()
    
    # Save to file (optional)
    safe_name = name.replace(' ', '_').replace('/', '_')
    filename = f'engineer_profile_{safe_name}.png'
    plt.savefig(filename, dpi=150, bbox_inches='tight')
    print(f"  âœ“ Created chart for {name}")
    
    plt.show()
    plt.close()

# Generate charts for all engineers
print("\n" + "="*80)
print("GENERATING INDIVIDUAL ENGINEER PROFILES")
print("="*80 + "\n")

for idx, row in df_filtered.iterrows():
    create_individual_profile(row, all_workflows)

print("\n" + "="*80)
print(f"âœ“ Successfully generated {len(df_filtered)} individual profile charts")
print("="*80 + "\n")

# Print summary statistics
print("\nSUMMARY STATISTICS:")
print(f"Total engineers analyzed: {len(df_filtered)}")
print(f"Total unique workflows identified: {len(all_workflows)}")
print(f"\nWorkflows identified:")
for i, wf in enumerate(all_workflows, 1):
    count = sum(1 for _, row in df_filtered.iterrows() 
                if pd.notna(row[workflow_col]) and wf in row[workflow_col])
    pct = (count / len(df_filtered) * 100)
    print(f"  {i}. {wf}: {count} engineers ({pct:.0f}%)")
```
